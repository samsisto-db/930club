```python
#hit urls gathered from google search returns.
#based on formatting of the url scrape HTML and get dates and status of the shows
#iloc gives us data from the specified row num (0) and desired column (urls)

t = showinfo_subset['urls']
#t = showinfo_read['urls']
t

status=[]
dates=[]

for i in t:
    if i[:12] == 'www.930.com/':
        if i[:18] == 'www.930.com/event/':
            i = str('http://')+i+str("?")
            r = requests.get(i)
            b = BeautifulSoup(r.text, 'html.parser')
            x = b.find('h3', { "class" : "price-range" })
            item = x.nextSibling
            print i
            if ('class' in item.attrs and 'custom' in item.attrs['class']) or (item.text == 'Sold Out'):
                status.append(1)
            else:
                status.append(0)
            
            q = b.find('h2', { "class" : "dates" })
            if (len(q.text) > 0):
                dates.append(q.text)
            else:
                dates.append('None found')
        
        elif i.find("/",12) == 15:
            i = 'http://www.930.com/event'+i[i.find("/",12):]+str("?")
            r = requests.get(i)
            b = BeautifulSoup(r.text, 'html.parser')
            x = b.find('h3', { "class" : "price-range" })
            item = x.nextSibling   
            if ('class' in item.attrs and 'custom' in item.attrs['class']) or (item.text == 'Sold Out'):
                status.append(1)
            else:
                status.append(0)
            
            q = b.find('h2', { "class" : "dates" })
            if (len(q.text) > 0):
                dates.append(q.text)
            else:
                dates.append('None found')
        
        else:
            status.append('Bad URL')
            dates.append('None found')
            continue
    else:
        status.append('Bad URL')
        dates.append('None found')
```
